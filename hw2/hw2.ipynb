How did lowering the confidence threshold lead to this change? Hint: Reference the Simpson’s Paradox section of the text.

<b>Result:</b>

![title](img/problem2a.png)

<i>Association Rules with Support Threshold = 10%, Confidence = 45%</i>

With support=10% and confidence=90%, we observe that only Opera Cake is in the Antecedent with Cherry Tart and yields only one association rule involving Cherry Tart in the Antecedent:

{Opera Cake, Cherry Tart} -> {Appricot Danish}

With support=10% and confidence=45%, we observe that Cherry Tart now appears without another item in the Antecedent producing the rules:

{Cherry Tart} -> {Opera Cake}

{Cherry Tart} -> {Apricot Danish}

An additional consequent, Opera Cake, is now observed when confidence=45%

A lower confidence threshold of 45% can be generalized as follows. If we were to repeat our procedure for finding associations on multiple samples of data, the <i>calculated</i> cofidence interval, which varies for each sample and is not the same as the <i>overall</i> confidence interval, will contain the given association rule 45% of the time. If an association occurs that lies outside the calculated confidence interval, then there was <= 65% chance of that association being made. Lowering the confidence interval allowed an association rule to be revealed that wouldn't make the cut when working with a 90% confidence interval. This relates closely with the Simpson's Paradox.

Load the Extended Bakery dataset (75000-out2-binary.csv) into Python using a Pandas dataframe. Calculate the binary correlation coefficient Φ for the Chocolate Coffee and Chocolate Cake items. Show whether the two items are symmetric binary variables via their co-presence and co-absence. Would an as- sociation rule between these items as antecedent and consequent have a high confidence level? Why or why not?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import matthews_corrcoef
%matplotlib inline
```

```python
df = pd.read_csv('dataset/75000-out2-binary.csv')
df.head(5)
```

```python
df.describe()
```

```python
# creates a confusion matrix
def confusion_matrix(x, y):
    matrix = np.array([[0, 0],
                       [0, 0]])
    for i in range(min(len(x), len(y))):
        matrix[x[i]][y[i]] += 1
    return matrix

# computes binary coefficient Phi
# Note: I also computed this using sklearn which produced a slightly different result
def compute_phi_coefficient(x, y):
    matrix = confusion_matrix(x, y)
    a = (matrix[1][1] * matrix[0][0]) - (matrix[1][0] * matrix[0][1])
    b = np.sqrt(sum(matrix[1]) * sum(matrix[0]) * sum(matrix[:, 0]) * sum(matrix[:, 1]))
    return np.divide(a, b)
```

```python
X = df['Chocolate Coffee']
Y = df['Chocolate Cake']

# using sklearn's 'Mathew's Correlation Coefficient' (same as phi)
matthews_corrcoef(X, Y)
```

```python
# using my function for computing phi
compute_phi_coefficient(X, Y)
```

```python
# copresence 
cm = confusion_matrix(X, Y)
total = min(len(X), len(Y))
copresence = cm[1][1] 
coabsence = cm[0][0]
diagonal = copresence + coabsence
print('Copresence ratio: {} / {} or {}%'.format(copresence, total, np.divide(copresence, total) * 100.))
print('Coabsence ratio: {} / {} or {}%'.format(coabsence, total, np.divide(coabsence, total) * 100.))
print('Total ratio falling along diagonal cells: {} / {} or {}%'.format(diagonal, total, np.divide(diagonal, total) * 100.))
```

<b>Result</b>

A generaliztion of phi coefficient on Wikipedia
(https://en.wikipedia.org/wiki/Phi_coefficient):

"Two binary variables are considered positively associated if most of the data falls along the diagonal cells."

We observed that 92.14% of the data was either copresent or coabsent. 

conf(Chocolate Cake implies Chocolate Coffee) = 3303 / 75000 or 4.404%

conf(not Chocolate Cake implies not Chocolate Coffee) = 65802 / 75000 or 87.736%

So we can create the association rule {not Chocolate Cake} -> {not Chocolate Coffee} from the 87.736% calculated confidence, but would not want to create a rule for the copresent due it having only a 4.404% calculated confidence.
